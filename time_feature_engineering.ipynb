```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx

plt.style.use(matplotx.styles.ayu['light'])
```

```python
# Load Dataset
fm_data = pd.read_csv("fm_data.csv")

fm_data.head()
```

```python
# Remove unnecessary column
fm_data = fm_data.drop(columns=["Unnamed: 0"])
```

```python
# Combine Date and Time into one column
fm_data["datetime"] = pd.to_datetime(fm_data["Date"] + " " + fm_data["Time"])
# Extract Hour
fm_data["hour"] = fm_data["datetime"].dt.hour
```

```python
# Create Time Period Feature
def get_time_period(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"

fm_data["time_period"] = fm_data["hour"].apply(get_time_period)
```

```python
# Extract Day of Week
fm_data["day_of_week"] = fm_data["datetime"].dt.day_name()
```

```python
# Weekend Feature
fm_data["is_weekend"] = fm_data["day_of_week"].isin(["Saturday", "Sunday"])
```
